```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://isimsf.rnu.tn/fra/enseignants"
```

```python
response = requests.get(url)
```

```python
response
```

```python
data = response.text
```

```python
soup = BeautifulSoup(data,'html.parser')
```

```python
profiles = soup.find_all('div',{'class':'tcard-front'})
```

```python
# Initialize an empty list to store the extracted information.
l = []

# Loop through the profiles.
for profile in profiles:
    # Extract the name from a <div> element with class 'h2'.
    name = profile.find('div', {'class': 'h2'}).text.strip()
    print(name)

    # Extract the email address.
    mail = profile.find_all('a')[1]
    address = 'None'  # Default value in case email address is not found.
    if len(mail.get('href').strip()) > 7:
        address = mail.get('href')[7:]
    print(address)

    # Extract the web page URL.
    web_page = profile.find_all('a')[0].get('href')
    print(web_page)

    # Append the extracted information as a list to the 'l' list.
    l.append([name, address, web_page])

    print('*****************')
```

```python
import pandas as pd
df = pd.DataFrame(l,columns=['name','email','webpage'])
```
